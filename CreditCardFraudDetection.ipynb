# Splitting Data

```python
# separating the X and Y vales
# X-->Input parameters
# Y--> Output Value
X=df1.drop(['Class'],axis=1)
Y=df1["Class"]
```

```python
X.head()
```

```python
Y.head()
```

```python
print(X.shape)
print(Y.shape)
```

```python
# converting these values into array format
X=X.values
Y=Y.values
```

# // Traninig and Testing Data Bifurcation

```python
# using scikit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest=train_test_split(X,Y,test_size=0.2,random_state=42)
```

# // Buiding DIfferent models

# 1) Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(xTrain,yTrain)
```

```python
# predictions
yPred=rfc.predict(xTest)
yPred
```

# // Building Evaluating Parameters

```python
# Evaluating the Classifier
# Checking Every score of Classifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import f1_score,matthews_corrcoef
from sklearn.metrics import confusion_matrix
```

```python
n_outliers=len(fraud_trasc)
n_outliers
```

```python
n_errors=(yPred != yTest).sum()
n_errors
```

```python
print("The model used is Random Forest classifier")
  
acc = accuracy_score(yTest, yPred)
print("The accuracy is {}".format(acc))
prec = precision_score(yTest, yPred)
print("The precision is {}".format(prec))
  
rec = recall_score(yTest, yPred)
print("The recall is {}".format(rec))
  
f1_Score = f1_score(yTest, yPred)
print("The F1-Score is {}".format(f1_Score))
  
MCC = matthews_corrcoef(yTest, yPred)
print("The Matthews correlation coefficient is{}".format(MCC))
```
